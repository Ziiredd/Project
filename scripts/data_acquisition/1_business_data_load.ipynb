### **1.2. Alcance**
---

Se propone la implementación de un modelo de aprendizaje profundo que permita realizar predicciones precisas en las series de tiempo que corresponden al histórico de consumo de los clientes para cada uno de los productos del portafolio. El modelo se basará en redes neuronales profundas y se entrenará con conjuntos de datos masivos para mejorar su capacidad de generalización y precisión en la predicción. Los datos de entrada incluirán información de la demanda histórica de cada producto, así como también información de variables externas que puedan influir en el comportamiento del mercado, como cambios estacionales, eventos especiales y tendencias del mercado. La precisión del modelo será evaluada mediante medidas estadísticas como el error medio absoluto (MAE) y el error medio cuadrático (MSE), y se realizarán ajustes en el modelo según sea necesario para mejorar su desempeño.

Los beneficios para la compañía serán la mejora en la toma de decisiones en cuanto al abastecimiento estratégico de su portafolio, reducción de costos y mayor eficiencia en el proceso logístico, lo que contribuirá a mejorar su flujo de caja y capital de trabajo.

### **1.3. Plan**
---

1. Recopilación de datos: El primer paso será recopilar todos los datos históricos de consumo de los clientes para cada uno de los productos del portafolio de la compañía. También se deberán obtener los datos de variables externas relevantes, como la estacionalidad, la demanda de la industria, el comportamiento del mercado, entre otros.

2. Preprocesamiento de datos: Una vez que se tenga toda la información recolectada, se deberán realizar tareas de preprocesamiento de datos para prepararlos para su uso en el modelo. Esto incluye la limpieza, normalización y transformación de los datos.

3. Selección de variables: A continuación, se deberán seleccionar las variables más relevantes para el modelo. Se pueden utilizar técnicas como análisis de correlación y modelos estadísticos para determinar qué variables tienen mayor impacto en la demanda de los productos.

4. Diseño del modelo: En esta etapa se deberá diseñar el modelo de redes neuronales recurrentes que permita realizar predicciones precisas en las series de tiempo de cada uno de los productos del portafolio de la compañía. Se deben considerar diferentes arquitecturas, hiperparámetros y técnicas de regularización para obtener el mejor rendimiento.

5. Entrenamiento del modelo: Una vez que se tenga el modelo diseñado, se procederá a entrenarlo utilizando los datos históricos previamente preprocesados. Se utilizarán técnicas de validación cruzada y ajuste de hiperparámetros para asegurarse de que el modelo esté bien ajustado y generalice adecuadamente.

6. Validación del modelo: Después de entrenar el modelo, se deberá realizar una validación exhaustiva del mismo para evaluar su rendimiento en datos no vistos. Se pueden utilizar diferentes métricas de evaluación, como el error cuadrático medio, el coeficiente de determinación y otras métricas de precisión.

7. Implementación del modelo: Finalmente, una vez que el modelo haya sido validado y se haya comprobado su eficacia, se deberá implementar en producción para ser utilizado por el área de planeación de demanda y suministro de la compañía. Se puede crear una interfaz de usuario para facilitar su uso y permitir que los usuarios puedan realizar predicciones de demanda para diferentes productos y en diferentes periodos de tiempo.
